```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Feb  6 15:04:02 2019

@author: regisferreiradasilva
"""

from sklearn import linear_model

import matplotlib.pyplot as plt
import numpy as np
import random
from scipy import optimize
import csv

#-----------------------------------------------------------#
#Etape 1 creation des variables

Tableau = []
t=[]
x = []
y = []
vx=[]
vy=[]
z=[]

#variables modelisation polynomiale, coefficients a modifier
a=0
b=1
c=0
d=0

#declaration de fonction polynomiale modelisation
def f(t):
    return a+b*t+c*t**2+d*t**3

#-----------------------------------------------------------#
#Etape 2 conversion fichier csv

g= open('mt_rectiligne_uni.csv','r')
csv_g = csv.reader(g)

i=0
for row in csv_g :
        Tableau.append(row)
        print(i," -----> ",row)
        i=i+1
g.close

#-----------------------------------------------------------#
#Etape 3 remplissage tableau par les valeurs experimentales
n = len(Tableau) 
i=2    #################  on commence à la 3ème ligne
while i<n-1: ##### on arrête à l'avant dernière
# on remplace la virgule par un point et on convertit en float
    t.append(float(Tableau[i][0].replace(',','.'))) 
    x.append(float(Tableau[i][1].replace(',','.')))
    y.append(float(Tableau[i][2].replace(',','.')))
   # vx.append(float(Tableau[i][3].replace(',','.')))
    #vy.append(float(Tableau[i][4].replace(',','.')))
  #  z.append(f(t[i][0]))
    i=i+1

#----------------------------------------------------------------------------------------#
# Step 1: training data

def test_func(x, a, b):
    return a * x +b

params, params_covariance = optimize.curve_fit(test_func, x, y,
                                               p0=[2, 2])

print(params)



plt.scatter(x,y,label='Donnees')
plt.plot(x, test_func(x, params[0], params[1]),
         label='Fitted function')

#----------------------------------------------------------------------------------------#
# Step 2: define and train a model

model = linear_model.LinearRegression()
model.fit(x, y)
#
#print(model.coef_, model.intercept_)

##----------------------------------------------------------------------------------------#
## Step 3: prediction
#
#x_new_min = 0.0
#x_new_max = 10.0
#
#X_NEW = np.linspace(x_new_min, x_new_max, 100)
#X_NEW = X_NEW[:,np.newaxis]
#
#Y_NEW = model.predict(X_NEW)
#
#plt.plot(X_NEW, Y_NEW, color='coral', linewidth=3)
#
#plt.grid()
#plt.xlim(x_new_min,x_new_max)
#plt.ylim(0,10)

plt.title("Simple Linear Regression using scikit-learn and python 3",fontsize=10)
plt.xlabel('x')
plt.ylabel('y')

plt.savefig("simple_linear_regression.png", bbox_inches='tight')
plt.show()
```
